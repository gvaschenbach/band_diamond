```r
library( tidyverse )
library( broom )
library( ggvis )
```

```r
str( diamonds )
```

```r
diamonds %>%
    ggvis( ~price, ~carat ) %>%
    layer_bars()
```

```r
ggplot( diamonds, aes( x = log( price ) ) ) +
    geom_density() +
    theme_classic()
```

```r
fit.lm <- lm( price
            ~ poly( carat, 3 )
            + cut
            + clarity 
            , data = diamonds )
```

```r
summary( fit.lm )
```

```r
tidy( fit.lm )
```

```r
df <- diamonds
```

```r
df$predicted <- predict( fit.lm )
```

```r
ggplot( df ) +
    geom_point( aes( x = price, y = predicted ) ) +
    theme_classic()
```

```r
ggplot( df, aes( x = carat, y = price ) ) +
    geom_point() +
    geom_smooth( se = FALSE ) +
    theme_classic()
```

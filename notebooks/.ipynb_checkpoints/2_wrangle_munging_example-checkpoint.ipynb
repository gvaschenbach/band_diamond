### Dealing with missing data exersize

```r
df.d <- readxl::read_excel( '../data/raw/dealing-with-missing-data.xlsx' )
```

```r
head( df.d )
```

```r
names( df.d ) <- c( 'client_id', 'average_tickets_no', 'employee_no', 'contract_value_no', 'industry_txt' )
```

### After removing the rows with missing data, how many rows are left?

```r
nrow( df.d[complete.cases( df.d),] )
```

### What is the impact of deleting the rows with missing data on the slope of the line?

```r
summary( lm( df.d$average_tickets_no ~ df.d$employee_no ) )
```

```r
ggplot( df.d, aes( x = employee_no, y = average_tickets_no ) ) +
    geom_point() +
    geom_smooth( se = FALSE, method = 'lm') +
    theme_classic()
```

```r
summary( lm( df.d[complete.cases(df.d),]$average_tickets_no ~ df.d[complete.cases(df.d),]$employee_no ) )
```

```r
ggplot( df.d[complete.cases( df.d ),], aes( x = employee_no, y = average_tickets_no ) ) +
    geom_point() +
    geom_smooth( se = FALSE, method = 'lm') +
    theme_classic()
```

## Imputation

```r
df.d[is.na( df.d$average_tickets_no ),]$average_tickets_no <- mean( df.d$average_tickets_no, na.rm = TRUE )
```

```r
head( df.d )
```

```r
summary( lm( df.d$average_tickets_no ~ df.d$employee_no ) )
```

```r
ggplot( df.d, aes( x = employee_no, y = average_tickets_no ) ) +
    geom_point() +
    geom_smooth( se = FALSE, method = 'lm' ) +
    theme_classic()
```
